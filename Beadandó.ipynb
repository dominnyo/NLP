```python
import pandas as pd

df = pd.read_parquet("hf://datasets/papermoon98/comment_toxicity/data/train-00000-of-00001.parquet")
```

```python
# prompt: list the first 10 rows of the df dataset

df.head(10)
```

```python
def clean_text(text):
  text = str(text) # Ensure the input is a string
  if text.startswith("Comment:"):
    text = text[len("Comment:"):]
  if text.endswith("Toxicity Labels:"):
    text = text[:-len("Toxicity Labels:")]
  return text.strip()

# Change 'text' to 'formatted_text' to match the actual column name
df['formatted_text'] = df['formatted_text'].apply(clean_text)
df.head()
```

```python
import re


def clean_text(text):
  text = re.sub(r"\\n|\\t|\\r", " ", text)  # Replace \n, \t, \r with spaces
  text = re.sub(r"\\", "", text) #remove remaining backslashes
  return text.strip()

df['formatted_text'] = df['formatted_text'].apply(clean_text)
df.head()
```

```python
import re

def clean_text(text):
  text = re.sub(r"[^a-zA-Z\s]", "", text) #ami nem betű, azt space-re cseréli
  return text.strip()  #eltünteti a felesleges white spacekat

df['formatted_text'] = df['formatted_text'].apply(clean_text)
df.head()
```

```python
def clean_text(text):
  text = text.lower()#kisbetűre alakít mindent
  return text.strip()

df['formatted_text'] = df['formatted_text'].apply(clean_text)
df.head()
```

```python
!pip install nltk

import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = text.split()
    lemmatized_words = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(lemmatized_words)

df['formatted_text'] = df['formatted_text'].apply(lemmatize_text)
df.head()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    return " ".join(filtered_words)

df['formatted_text'] = df['formatted_text'].apply(remove_stopwords)
df.head()
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

# Apply word tokenization to the 'formatted_text' column
df['formatted_text'] = df['formatted_text'].apply(word_tokenize)

df.head()
```

```python
# prompt: irja ki az első 10 sort a df adatbázisból

df.head(10)
```

```python
# prompt: split the whole dataset into 80% training 10% validation and 10% testing

from sklearn.model_selection import train_test_split

# Assuming 'df' is your DataFrame with 'formatted_text' and your target variable
# Replace 'target_variable' with the actual name of your target column

# Split the data into training and temporary sets (80% train, 20% temp)
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42) #random_state for reproducibility

# Split the temporary set into validation and testing sets (50% validation, 50% testing)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print("Training set size:", len(train_df))
print("Validation set size:", len(val_df))
print("Testing set size:", len(test_df))
```

```python
# prompt: do vectorization on the database

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming 'df' is your DataFrame with the 'formatted_text' column

# Initialize the TF-IDF vectorizer
vectorizer=TfidfVectorizer(tokenizer=lambda doc: doc,preprocessor=lambda doc:doc, lowercase=False,max_features=10000)
# Fit and transform the training data
train_vectors = vectorizer.fit_transform(train_df['formatted_text'])

# Transform the validation and testing data using the same vectorizer
val_vectors = vectorizer.transform(val_df['formatted_text'])
test_vectors = vectorizer.transform(test_df['formatted_text'])

# Now you have sparse matrices 'train_vectors', 'val_vectors', and 'test_vectors'
# that represent your text data as numerical vectors

print(train_vectors.shape)
print(val_vectors.shape)
test_vectors.shape
```

```python
# prompt: unsparse all vectors

train_vectors = train_vectors.toarray()
val_vectors = val_vectors.toarray()
test_vectors = test_vectors.toarray()

print(train_vectors.shape)
print(val_vectors.shape)
test_vectors.shape
```

```python
# prompt: convert each label into a single digit based on if it has any true values

def convert_labels(labels_map):
  """Converts a map of toxicity labels to a single digit."""
  # Check if any of the values in the labels_map are True
  # and return 1 if so, otherwise return 0.
  return 1 if any(labels_map.values()) else 0

# Apply the function to each DataFrame
train_df['toxicity_labels'] = train_df['toxicity_labels'].apply(convert_labels)
val_df['toxicity_labels'] = val_df['toxicity_labels'].apply(convert_labels)
test_df['toxicity_labels'] = test_df['toxicity_labels'].apply(convert_labels)
```

```python
import numpy as np
```

```python
# prompt: export obove numbers into np arraz

train_toxicity_labels = np.array(train_df['toxicity_labels'].tolist())
val_toxicity_labels = np.array(val_df['toxicity_labels'].tolist())
test_toxicity_labels = np.array(test_df['toxicity_labels'].tolist())
```

```python
# prompt: make a tensorflow model for text classificatyion model

import tensorflow as tf

# Define the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(train_vectors.shape[1],)),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer with sigmoid for multi-label classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',  # Use binary_crossentropy for multi-label classification
              metrics=['accuracy'])

# Train the model
model.fit(train_vectors, train_toxicity_labels, epochs=11, batch_size=500, validation_data=(val_vectors, val_toxicity_labels))


# Evaluate the model
loss, accuracy = model.evaluate(test_vectors, test_toxicity_labels)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```
